```python
# Import cartopy.crs to provide classes for defining coordinate system
import cartopy.crs as ccrs
# Import matplotlib.pylot to be used for plotting and creating visualizations
import matplotlib.pyplot as plt

# Create a new figure object from matplotlib.pylot that should be 10 inches wide and 5 inches tall
fig = plt.figure(figsize=(10, 5))
# Add a 1x1 grid subplot to the figure and create a GeoAxes object using PlateCarree
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
# Add coastlines on the map
ax.coastlines()
# Add a title to the plot Map of the Earth
plt.title('Map of the Earth')
# Show the created map
plt.show()
```

In the following cell, add an arrow saying "We are here" that is pointing to Charlotte North Carolina


```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Charlotte's coordinates
charlotte_lon = -80.8431
charlotte_lat = 35.2271

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

#Add an arrow saying "We are here" that is pointing to Charlotte North Carolina
ax.annotate(
    "We are here",
    xy=(charlotte_lon, charlotte_lat),  # Coordinates of the arrow's tip
    xytext=(charlotte_lon - 15, charlotte_lat + 5),  # Coordinates of the text
    arrowprops=dict(facecolor="red", width=1, headwidth=8), #Arrow style
    color = 'blue', #Text color
    fontsize = 12, #Text size
    horizontalalignment = 'right'
)

plt.title('Map of the Earth')
plt.show()
```

Now add major ocean labels

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Charlotte's coordinates
charlotte_lon = -80.8431
charlotte_lat = 35.2271

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

# Add the arrow and text
ax.annotate(
    "We are here",
    xy=(charlotte_lon, charlotte_lat),  # Coordinates of the arrow's tip
    xytext=(charlotte_lon - 15, charlotte_lat + 5),  # Coordinates of the text
    arrowprops=dict(facecolor="red", width=1, headwidth=8), #Arrow style
    color = 'blue', #Text color
    fontsize = 12, #Text size
    horizontalalignment = 'right')

# Add major ocean labels (Atlantic, Pacific, etc.)
ax.text(-45, 30, 'Atlantic Ocean', fontsize=12, color='navy', ha='center', va='center')
ax.text(-130, 0, 'Pacific Ocean', fontsize=12, color='navy', ha='center', va='center')
ax.text(80, -20, 'Indian Ocean', fontsize=12, color='navy', ha='center', va='center')
ax.text(10, 75, 'Arctic Ocean', fontsize=12, color='navy', ha='center', va='center')
ax.text(-70, -60, 'Southern Ocean', fontsize=12, color='navy', ha='center', va='center')

plt.title('Map of the Earth')
plt.show()
```
